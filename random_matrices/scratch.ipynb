```python
from scipy.linalg import solve
# Function to get trophic levels (for alpha calculation)
def getTrophicLevels(adj):
    flws = adj.copy()
    for to in range(len(flws)):
        totflow=sum(flws[:,to])
        if totflow > 0:
            flws[:,to]=flws[:,to]/totflow
        flws[to,to]=-1
    return solve(flws.T,-1.*np.ones(len(flws)))-1.
```

```python
def calc_msf(job):
    sp = job.sp

    # Load kappa and adj from shared data
    sd_fn = project.fn('shared_data.h5')
    with signac.H5Store(sd_fn).open(mode='r') as sd:
        kap_arr = np.array(sd['kappa'])
        adj = np.array(sd['adj_mats'])[sp['wi']]
    
    # Load P matrix
    with job.stores.P_matrix:
        P = np.array(job.stores.P_matrix.P)
    
    # Make C matrix
    C = np.zeros(adj.shape)
    C[sp.di][sp.di] = sp.delta

    # Construct msf and write to file
    msf = np.empty(len(kap_arr))
    for k, kap in enumerate(kap_arr):
        J = P - kap*C
        J_evals, Jevecs = np.linalg.eig(J)
        msf[k] = max(J_evals.real)
    
    return msf
```

```python
def heatmap(data, row_labels, col_labels, 
            ax, cbarlabel="", **kwargs):
    im = ax.imshow(data, **kwargs)
    cbar = ax.figure.colorbar(im, ax=ax, location="right", shrink=0.8)
    cbar.ax.set_ylabel(cbarlabel, rotation=-90, fontsize=15, labelpad=20)
    ax.set_xticks(np.arange(data.shape[1]))
    ax.set_xticklabels(col_labels, fontsize=25)
    ax.set_yticks(np.arange(data.shape[0]))
    ax.set_yticklabels(row_labels, fontsize=25)
    ax.spines[:].set_visible(False)
    ax.set_xticks(np.arange(data.shape[1]+1)-.5, minor=True)
    ax.set_yticks(np.arange(data.shape[0]+1)-.5, minor=True)
    ax.grid(which="minor", color="black", linestyle='-', linewidth=3)
    ax.tick_params(which="minor", bottom=False, left=False)
```

```python
from itertools import product
from matplotlib import pyplot as plt
from matplotlib import cm
from matplotlib import colors

webnms = ["Omnivory","Exploitative","Chain","Apparent"]
specnms = ["r","n","p"]

ddi_mat = []
ddi_mat_diff = []
for di in np.flip(list(schema['di'][int])):
    di = int(di)
    temp = []
    temp_diff = []
    for wi in list(schema['wi'][int]):
        print('di={}, wi={}'.format(di,wi))
        # Find the ddi fraction with sigma unconstrained
        tot = len(project.find_jobs({'wi': wi, 'di': di}))
        print('num unconstrained samples:',tot)
        num_ddi = len(project.find_jobs({'wi': wi, 'di': di, 
                                 'doc.nonspatial_stable': True, 
                                 'doc.msf_crosseszero': True}))
        ddi_frac = num_ddi / tot
        temp.append(ddi_frac)
        print('uncon:',ddi_frac)
        
        # Find the ddi fraction with difference in sigma constrained above some threshold
        thresh = 0.5
        tot = len(project.find_jobs({'wi': wi, 'di': di, 
                                     'doc.gamma_diff': {'$gt': thresh}}))
        print('num constrained samples:',tot)
        num_ddi = len(project.find_jobs({'wi': wi, 'di': di, 
                                 'doc.nonspatial_stable': True, 
                                 'doc.msf_crosseszero': True, 
                                 'doc.gamma_diff': {'$gt': thresh}}))
        ddi_frac_constrained = num_ddi / tot if tot>0 else 0
        #Show difference as zero, should really just cut this column out
        ddi_diff = ddi_frac - ddi_frac_constrained if tot>0 else 0
        temp_diff.append(ddi_diff)
        print('con:',ddi_frac_constrained)
        print('diff:', ddi_diff)
        print('---')
    ddi_mat.append(temp)
    ddi_mat_diff.append(temp_diff)
ddi_mat = np.array(ddi_mat)
ddi_mat_diff = np.array(ddi_mat_diff)
# Plot unconstrained
fig, ax = plt.subplots(figsize=(12, 9))
heatmap(ddi_mat, np.flip(specnms), webnms, ax=ax, 
        cbarlabel="Turing Fraction", cmap="Blues")
# Plot constrained
bound = max([abs(np.min(ddi_mat_diff)),abs(np.max(ddi_mat_diff))])
norm = colors.TwoSlopeNorm(vmin=-bound, vcenter=0, vmax=bound)
fig, ax = plt.subplots(figsize=(12, 9))
heatmap(ddi_mat_diff, np.flip(specnms), webnms, ax=ax, 
        cbarlabel="Delta Turing Fraction", cmap="RdBu", norm=norm)
```
